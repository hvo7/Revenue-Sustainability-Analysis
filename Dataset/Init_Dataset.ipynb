### Initialize the dataset into the raw_data file

```python
import pandas as pd
import numpy as np
import kaggle
from kaggle.api.kaggle_api_extended import KaggleApi
import os
from pathlib import Path
api = KaggleApi()
api.authenticate()
api.model_list_cli()

base_dir = Path('c:\\Users\\henry\\OneDrive\\Personal Career\\Personal Projects\\GitHub\\Revenue-Sustainability-Analysis')

raw_data_dir = Path(base_dir / 'raw_data')
if not raw_data_dir.exists():
    os.mkdir(raw_data_dir)

# Store purely kaggle data in the Kaggle folder
os.chdir(raw_data_dir)
kaggle_dir = (raw_data_dir / 'Kaggle')
if not kaggle_dir.exists():
    os.mkdir(kaggle_dir)
```

```python
datasets = [
        "rivalytics/saas-subscription-and-churn-analytics-dataset"
]

for d in datasets:
    api.dataset_download_files(
        d,
        path="Kaggle",
        unzip=True
    )
```
